```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
state_df=pd.read_csv("state_wise_daily.csv")
```

```python
state_df.info()
```

```python
state_df.head()
```

```python
state_df.isnull()
```

```python
state_df.loc[[237,326],['DL']]
```

```python
june_month=state_df[237:327]
june_month.isnull()
```

```python
dates=np.array(june_month['Date'].drop_duplicates())
dates
```

```python
DL=state_df['DL']
DL.isnull()
```

```python
confirmed_case=DL[237:327:3].to_numpy()
```

```python
recovered_case=DL[238:327:3].to_numpy()
```

```python
deceased_case=DL[239:327:3].to_numpy()
```

```python
DL.to_numpy()
```

```python
date=np.arange(len(dates))
date
```

```python
style.use('ggplot')
plt.figure(figsize=(25,10))
plt.xticks(date,dates,rotation=30,fontsize=15)
plt.yticks(fontsize=15)
width=0.3
plt.bar(date,confirmed_case,color='red',label='confirmed',width=0.3)
plt.bar(date+width,recovered_case,color='green',label='recovered',width=0.3)
plt.bar(date+width+width,deceased_case,color='grey',label='deceased',width=0.3)
plt.title('covid19 Report of Delhi june2020',fontsize=25)
plt.xlabel('Month of June 2020',fontsize=20)
plt.ylabel('CASES',fontsize=20)
plt.legend(fontsize=15)
#plt.grid(True,color='black')
plt.savefig('covid19 Report of Delhi june2020 BAR GRAPH.png')
plt.show()
```
